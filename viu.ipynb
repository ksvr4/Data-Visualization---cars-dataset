```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(r"C:\Users\ksvra\Downloads\car.csv")
print(data)
```

```python
def count_car(name,i=0):
    for car in data['make'] :
        if car == name :
            i = i+1
    return i
car = ['BMW','Kia','Volvo','Nissan','Chevrolet','Audi']
no_car = []
for c in car :
    no_car.append(count_car(c))
k = 0
for i in car :
    plt.bar([i], [no_car[k]], label=i, width=.5)
    k+=1
print("Number of cars are : ")
print(no_car)
plt.legend()
plt.xlabel("CARS")
plt.ylabel("No_of_sales")
plt.title("Sales of each Brand")
plt.show()
```

```python
def count_car(name,i=0):
    for car in data['make'] :
        if car == name :
            i = i+1
    return i
def count_body(name,b=0):
    for body in data['body'] :
        if body == name:
                b = b+1
    return b
car = ['BMW','Kia','Volvo','Nissan']
body = ['SUV','Sedan','Convertible','Coupe']
no_of_car = []
for car in car :
    no_of_car.append(count_car(car))
no_of_body = []
for body in body :
    no_of_body.append(count_body(body))
all = [no_of_car,no_of_body]
print ("Number of cars are ",no_of_car)
print("Number of body ",no_of_body)
plt.boxplot(all,labels=['No of cars','No of body'],showmeans=True)
plt.xlabel("CARS")
plt.ylabel("No_of_sales")
plt.title("Box plot of Number of cars and Number of Body sold ")
plt.show()
```

```python
area_data=pd.read_csv(r'C:\Users\ksvra\Downloads\car.csv',usecols=['year','make'])
data_2015=area_data[area_data['year'] == 2015]
data_2014=area_data[area_data['year'] == 2014]
data_2013=area_data[area_data['year'] == 2013]
data_2012=area_data[area_data['year'] == 2012]
year = [2015,2014,2013,2012]
def count_car(name,year,i=0):
    if year == 2015:
        for car in data_2015['make'] :
            if car == name :
                i = i+1
    elif year == 2014:
        for car in data_2014['make'] :
            if car == name :
                i = i+1
    elif year == 2012:
                for car in data_2012['make']:
                    if car == name:
                        i = i + 1
    else:
        for car in data_2013['make'] :
            if car == name :
                i = i+1
    return i
bmw_sales = [count_car('BMW',2015),count_car('BMW',2014),count_car('BMW',2013),count_car('BMW',2012)]
audi_sales = [count_car('Audi',2015),count_car('Audi',2014),count_car('Audi',2013),count_car('Audi',2012)]
plt.scatter(year,bmw_sales,label='BMW sales',color = 'b')
plt.scatter(year,audi_sales,label='Audi sales',color = 'r')
print("BMW sales are : ",bmw_sales)
print("Audi sales are : ",audi_sales)
plt.xlabel('Years')
plt.ylabel('Sales')
plt.title("Scatter plot of sales in both BMW and Audi")
plt.legend()
plt.show()
```
